# Data Source Type

- First party data: Gathered Directly from inside your organization
- Second party data: Gathered from outside your organization but Directly from original source
- Third party data: Gathered from outside your organization and aggregated

# EDA

- Exploratory data analysis or EDA is the process of investigating, ​organizing and analyzing data sets. ​And summarizing their main characteristics often employing data wrangling and ​visualization methods. 
- ​The six main practices of EDA are discovering (understanding the data) , structuring (organizing and categorizing), ​cleaning (removing errors or noise), joining (adjusting data or context by adding new data or information), validating (Verifying the consistency of the data) and presenting (making the final dataset available for modeling and further steps). 
- Example steps:
    - Discovering: You check out the overall shape, size, and content of the dataset. You find it is short on data. 
    - Joining: You add more data.
    - Validating: You perform a quick check that the new data doesn’t have mistakes or misspellings. 
    - Structuring: You structure the data in different time periods and segments to understand trends. 
    - Validating: You do another quick check to ensure the new columns you’ve made in structuring are correctly designed. 
    - Cleaning: You check for outliers, missing data, and needs for conversions or transformations. 
    - Validating: After cleaning, you double check the changes you made are correct and accurate. 
    - Presenting: You share your dataset with a peer.
