# Comparator

- Boolean Operators: Compares 2 values and produces boolean values 0 or 1. 6 boolean Comparators: ==, >, <, >=, <=, !=
- Logical Operators: Connects multiple statements together and performs a more complex comparisons. 3 logical operators: and, or, not. For string comparison, it compares the first characters. if they tie, the next character is compared.
- Mathematical Operators: Used in mathematical expressions: +, -, /, *, **, //, % 

# Conditional Statements

- if, elif, else: It allows branching. Branching describes the ability of a program to alter its execution sequence.

### Example 1

```python
x = 8
if x > 5:
   print('x is greater than five')
elif x < 5:
   print('x is less than five')
else:
   print('x is equal to five')
```

```python
def greater_than_ten(x):
    if x > 10:
        return True
    return False

print(greater_than_ten(15))
print(greater_than_ten(2))
```
