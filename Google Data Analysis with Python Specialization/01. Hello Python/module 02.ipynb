# Jupyter Notebook Shortcuts

In **Jupyter Notebook**, there are **two modes** that determine how shortcuts work:

* **Command Mode** → for managing cells (press `Esc` to enter)
* **Edit Mode** → for editing code/text inside a cell (press `Enter`)

Below is a **large collection of useful keyboard shortcuts**.

---

1. Cell Creation & Deletion

*(Command Mode – press `Esc` first)*

| Shortcut    | Action                |
| ----------- | --------------------- |
| `A`         | Insert cell **above** |
| `B`         | Insert cell **below** |
| `DD`        | Delete selected cell  |
| `Z`         | Undo cell deletion    |
| `X`         | Cut cell              |
| `C`         | Copy cell             |
| `V`         | Paste cell **below**  |
| `Shift + V` | Paste cell **above**  |

---

2. Running Cells

| Shortcut        | Action                             |
| --------------- | ---------------------------------- |
| `Shift + Enter` | Run cell and move to next          |
| `Ctrl + Enter`  | Run cell and stay                  |
| `Alt + Enter`   | Run cell and insert new cell below |

---

3. Changing Cell Type

| Shortcut | Action                       |
| -------- | ---------------------------- |
| `Y`      | Convert to **Code** cell     |
| `M`      | Convert to **Markdown** cell |
| `R`      | Convert to **Raw** cell      |

---

4. Cell Navigation

| Shortcut     | Action             |
| ------------ | ------------------ |
| `Up Arrow`   | Select cell above  |
| `Down Arrow` | Select cell below  |
| `K`          | Move to cell above |
| `J`          | Move to cell below |

---

5. Moving Cells

| Shortcut           | Action                         |
| ------------------ | ------------------------------ |
| `Shift + Up`       | Select multiple cells upward   |
| `Shift + Down`     | Select multiple cells downward |
| `Shift + M`        | Merge selected cells           |
| `Ctrl + Shift + -` | Split cell at cursor           |

---

6. Notebook Operations

| Shortcut   | Action             |
| ---------- | ------------------ |
| `S`        | Save notebook      |
| `Ctrl + S` | Save notebook      |
| `I I`      | Interrupt kernel   |
| `0 0`      | Restart kernel     |
| `H`        | full key shortcut  |

---

7. Output Control

| Shortcut    | Action                  |
| ----------- | ----------------------- |
| `O`         | Toggle output           |
| `Shift + O` | Toggle output scrolling |
| `L`         | Toggle line numbers     |

---

8. Edit Mode Shortcuts

*(Press `Enter` to enter edit mode)*

| Shortcut           | Action                 |
| ------------------ | ---------------------- |
| `Ctrl + /`         | Comment/uncomment line |
| `Ctrl + Z`         | Undo                   |
| `Ctrl + Shift + Z` | Redo                   |
| `Ctrl + A`         | Select all             |
| `Ctrl + D`         | Delete line            |
| `Tab`              | Code completion        |
| `Shift + Tab`      | Show documentation     |

---

9. Useful Python/Jupyter Tricks

| Shortcut          | Action                          |
| ----------------- | ------------------------------- |
| `Tab`             | Auto-complete variable/function |
| `Shift + Tab`     | Show function help              |
| `?function_name`  | Show documentation              |
| `??function_name` | Show full source code           |

Example:

```python
import pandas as pd
pd.read_csv?
```

---

10. Markdown Shortcuts (inside Markdown cells)

| Shortcut    | Result        |
| ----------- | ------------- |
| `# Title`   | Heading 1     |
| `## Title`  | Heading 2     |
| `### Title` | Heading 3     |
| `**text**`  | Bold          |
| `*text*`    | Italic        |
| `- item`    | Bullet list   |
| `1. item`   | Numbered list |

# OOP Code Example

````python
class Spaceship:
   # Class attribute
   tractor_beam = 'off'

   # Instance attributes
   def __init__(self, name, kind):
       self.name = name
       self.kind = kind
       self.speed = None

  # Instance methods
   def warp(self, warp):
       self.speed = warp
       print(f'Warp {warp}, engage!')

   def tractor(self):
       if self.tractor_beam == 'off':
           self.tractor_beam = 'on'
           print('Tractor beam on.')
       else:
           self.tractor_beam = 'off'
           print('Tractor beam off')
````

````python
# Create an instance of the Spaceship class (i.e. "instantiate")
ship = Spaceship('Mockingbird','rescue frigate')

# Check ship's name
print(ship.name)

# Check what kind of ship it is
print(ship.kind)

# Check tractor beam status
print(ship.tractor_beam)
````

````python
# Set warp speed
ship.warp(7)

# Check speed
ship.speed
````

````python
# Toggle tractor beam
ship.tractor()

# Check tractor beam status
print(ship.tractor_beam)
````
