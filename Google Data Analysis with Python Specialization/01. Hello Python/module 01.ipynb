# Why Python

- Computer is designed with billions of transistors. These transistors either let electricity pass or block it. Using this duality, the computer thus can understand complex logic, but only in binary (either 0 or 1). But for human this is troublesome due to binary being a low level programming language, so high level language like python is created to mitigate this issue and as a bridge between human and computer
- It is open source, easy to learn, has huge community

# OOP

- Python is OOP language that lets it use both data and functionality that manipulates that data together
- Everything is object in python
- Allows us to pack information and ways to manipulate these information into one place.
- Class =  a class is an object’s data type that bundles data and functionality together. So it is possible to create bunch of tools into the object itself. A class is like a blueprint for all things that share characteristics and behaviors.
- Object is an instance of a class
- Method = a function that belongs to a class and performs certain actions or operations
- Attributes = Values associated with an object or class that are referenced by name with a dot notation. they dont use parethesis. Attributes allow us to access characteristics in a class without without doing anything to it or changing it
- Dot notation = how to access methods and attributes that belong to an object
