# Variables 

- Variables point to values. not the values themselves
- data type = attribute that describes a piece of data based on its values
- keywords cannot be used as variables
- dont use reserved names of existing functions
- letters, numbers and underscore
- cannot use parenthesis. and no special characters
- cannot start with a number. only start with letter or underscore
- Expressions: A combination of numbers, symbols, and variables to compute and return a result upon evaluation
- variables are containers and the values are what these contain.
