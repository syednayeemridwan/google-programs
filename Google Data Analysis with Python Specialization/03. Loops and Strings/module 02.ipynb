# For Loop

- In Python, a for loop is a piece of code that iterates over a sequence of values.
- A for loop is a control structure that allows you to execute a block of code the same number of times as there are elements in an iterable sequence (can be any iterable data type).
- use for loops ​when there's a sequence of elements ​that you want to iterate over.

```python
students = [['Igor', 'Sokolov'], ['Riko', 'Miyazaki'], ['Tuva', 'Johansen']]
for student in students:
   for name in student:
       print(name)
   print()
```
