# Indexing

- Indexing is Python's way ​of letting us refer to individual items ​within an iterable by their relative position

```python
x = 'values'
y = 100

print('''String formatting lets you insert {} into strings.
They can even be numbers, like {}.'''.format(x, y))
```

```python
var_a = 'A'
var_b = 'B'
print('{a}, {b}'.format(b=var_b, a=var_a))
```

```python
var_a = 'A'
var_b = 'B'
print('{1}, {0}'.format(var_a, var_b))
print('{0}, {1}'.format(var_a, var_b))
```

```python
# print('{}, {}, {}, {}, {}, {} ...'.format(1, 2, 3, 4, 5, 6))

print('{0}{1}{0}'.format('abra', 'cad'))
```

```python
var_a = 1
var_b = 2
print(f'{var_a} + {var_b}')
print(f'{var_a + var_b}')
print(f'var_a = {var_a} \nvar_b = {var_b}')
```

```python
num = 1000.987123
print(f'{num:.2f}')

num = 1000.987123
print(f'{num:.3e}')

decimal = 0.2497856
print(f'{decimal:.4%}')
```
