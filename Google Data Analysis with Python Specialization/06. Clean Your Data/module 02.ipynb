# Outliers

- global outliers (inaccuracies, extreme values): Should be thrown out
- contextual outliers (anomalies in terms of context): average height for mammal is outlier if the mammal is a mouse. 
- collective outliers (group of outliers that show pattern and are isolated from rest of the population)
- Double check if a data is outlier, and even if it is, should we exclude it.
